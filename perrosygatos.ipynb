```python
#1. Tracen un histograma del porcentaje de dueos de perros.

plt.style.use('seaborn')
plt.xlabel('Percentage of Dog Owners')
df['Percentage of Dog Owners'].hist(color='red', bins = 10, edgecolor = 'purple')
```

```python
#2. Tracen un histograma del porcentaje de dueos de gatos.

plt.xlabel('Percentage of Cat Owners')
df['Percentage of Cat Owners'].hist(color='orange', bins = 10, edgecolor = 'yellow')
```

De forma preliminar se puede concluir que al menos el 30% de los estados tiene gatos o perros. Sin embargo, se puede entender que hay mas personas con perros que con gatos.

```python
#3a. Qu podemos ver al comparar estos dos histogramas? Qu informacin nos dice?

plt.hist(df['Percentage of Cat Owners'], bins = 10, alpha = 0.8, edgecolor = 'black', label = 'Cats')
plt.hist(df['Percentage of Dog Owners'], color= 'brown', bins = 10, alpha = 0.8, edgecolor = 'black', label = 'Dogs')
plt.legend(loc = 'upper left')
```

Se puede concluir que aunque haya un porcentaje mayor de due;os de perros, quien tiene gatos tiende a tener mas dentro de la casa. Tienden a acumular mas.
